# Data Preparation and Cleaning

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Dataset

```python
df = pd.read_csv("C:\\Users\\rajso\\Downloads\\covid_19_india.csv")
```

# Make copy of Dataset

```python
df1 = df.copy()
```

# Reading Dataset

```python
df1.head()
```

# Checking data types,null values,unique value 

```python
df1.dtypes
```

```python
df1.describe()
```

```python
df1.info()
```

# Finding null value in the table

```python
df1.isnull()
```

```python
df1.isnull().sum()
```

```python
df1["State/UnionTerritory"].unique()
```

```python
df1["State/UnionTerritory"] = df1["State/UnionTerritory"].replace(['Bihar****','Telengana'],['Bihar','Telangana'])
```

```python
df1["State/UnionTerritory"].unique()
```

# Create seperate year column

```python
year = df1["Date"].apply(lambda x:x.split("-")[0])
```

```python
df1["Year"] = year
```

```python
df1
```

# Create seperate month column

```python

def Month(st):
    return st[5:7]
df1["Month"] = df1["Date"].apply(Month)
    
```

```python
df1
```

# Combine Union Territory Diu,Daman,Dadra,Nagar Haveli 

```python
df1[df1["State/UnionTerritory"]=="Daman & Diu"]
```

```python
df1[df1["State/UnionTerritory"]=="Daman and Diu"]
```

```python
df1[df1["State/UnionTerritory"]=="Dadra and Nagar Haveli"]
```

```python
df1[df1["State/UnionTerritory"]=="Dadra and Nagar Haveli and Daman and Diu"]
```

```python
df1["State/UnionTerritory"] = df1["State/UnionTerritory"].replace(['Daman & Diu','Daman and Diu','Dadra and Nagar Haveli'],['Dadra and Nagar Haveli and Daman and Diu','Dadra and Nagar Haveli and Daman and Diu','Dadra and Nagar Haveli and Daman and Diu'])
```

```python
df1["State/UnionTerritory"].unique()
```

# Combine Unassigned and Case being reassigned to states column

```python
df1[df1["State/UnionTerritory"]=="Unassigned"]
```

```python
df1["State/UnionTerritory"] = df1["State/UnionTerritory"].replace('Cases being reassigned to states','Unassigned')
```

# Create new column Month and Year

```python
month_1 = df1["Date"].apply(lambda x:x.split("-")[1])
year1 = df1["Date"].apply(lambda x:x.split("-")[0])

month_year =year1 + "-"+ month_1
month_year
```

```python
df1["M&Y"] = month_year
```

```python
new_df = df1.groupby("M&Y")["Cured","Deaths","Confirmed"].sum().sort_values("M&Y",inplace= False).reset_index()
```

```python
new_df
```

# Cured Cases in India by Line chart

```python
fig = plt.figure(figsize=(13,6))
plt.plot(new_df["M&Y"],new_df["Cured"])
plt.title("Cured Cases in India")
plt.xlabel('Month & Year')
plt.ylabel("Cured Case")
plt.tight_layout()
plt.show()
```

# Confirmed Cases in India by Line chart

```python
fig = plt.figure(figsize=(13,6))
plt.plot(new_df["M&Y"],new_df["Confirmed"],color='yellow')
plt.title("Confirmed Cases in India")
plt.xlabel('Month & Year')
plt.ylabel("Confirmed Case")
plt.tight_layout()
plt.show()
```

# Deaths Cases in India by Line chart

```python
fig = plt.figure(figsize=(13,6))
ax = plt.plot(new_df["M&Y"],new_df["Deaths"],color='r')

plt.title("Deaths Cases in India")
plt.xlabel('Month & Year')
plt.ylabel("Deaths Case")
plt.tight_layout()
plt.show()
```

# Deaths Cases in India by Bar chart

```python
fig = plt.figure(figsize=(13,6))
sns.set(style='darkgrid')
plt.tight_layout
ax = sns.barplot(new_df["M&Y"],new_df["Deaths"],color = 'r')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```

# Confirmed Cases in India by Bar chart

```python
fig = plt.figure(figsize=(13,8))
sns.set(style='darkgrid')
plt.tight_layout
ax = sns.barplot(new_df["M&Y"],new_df["Confirmed"],color = 'r')
ax.set_xticklabels(ax.get_xticklabels(), rotation=40, ha="right")
plt.show()
```
